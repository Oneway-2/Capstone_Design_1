```python
  
class SampleApp(tkinter.Tk):
    def __init__(self):
        print("SampleApp")
        tkinter.Tk.__init__(self)
        
        self.title('Helper')              # 창 제목 설정
        self.geometry("300x200+1550+50")   # 창 크기 (너비x높이+x좌표+y좌표)
        self.resizable(False, False)   # 창 크기 조절 여부
        
        self._frame = None
        self.switch_frame(StartPage)
        

    def switch_frame(self, frame_class):
        print("switch_frame")
        new_frame = frame_class(self)
        if self._frame is not None:
            self._frame.destroy()
        self._frame = new_frame
        self._frame.pack()
        


class StartPage(tkinter.Frame):  # 자동암기모드 (시험모드, 입력모드 둘 다 가능하게.)
    def __init__(self, master):
        print("StartPage")
        tkinter.Frame.__init__(self, master)        
        tkinter.Label(self, text="Start page", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tkinter.Button(self, text="Go to page one",
                  command=lambda: master.switch_frame(PageOne)).pack()
        tkinter.Button(self, text="Go to page two",
                  command=lambda: master.switch_frame(PageTwo)).pack()
        


class PageOne(tkinter.Frame):   # 시험모드 (자동암기모드로 나갈 수 있게)
    def __init__(self, master):
        print("PageOne")
        tkinter.Frame.__init__(self, master)
        tkinter.Frame.configure(self,bg='blue')
        tkinter.Label(self, text="Page one", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        tkinter.Button(self, text="Go back to start page",
                  command=lambda: master.switch_frame(StartPage)).pack()
        
            
class PageTwo(tkinter.Frame):  # 입력모드 (자동암기모드로 나갈 수 있게)
    def __init__(self, master):
        tkinter.Frame.__init__(self, master)

        
        print("PageTwo")
        meaning = ''
        spelling = ''
        
        ############################################:: Entry에 의해 실행되는 함수 ::###########################################  
        def save_mean(enter_value):   # 뜻을 입력받고 meaning 에 저장  # word_mean_enter Entry와 엮여있다.
            global meaning
            meaning = tkinter.Entry.get(word_mean_enter)
            check_mean.config(text='입력한 뜻 : ' + meaning)
            print(meaning)
            save_word.config(text = '저장')

    
        def save_spelling(enter_value):   # 스펠링을 입력받고 meaning 에 저장  # word_spelling_enter Entry와 엮여있다.
            global spelling
            spelling = tkinter.Entry.get(word_spelling_enter)
            check_spelling.config(text='입력한 스펠링 : ' + spelling)
            print(spelling)
            save_word.config(text = '저장')

        ###########################################:: Entry에 의해 실행되는 함수 ::############################################  



        ###########################################:: Button에 의해 실행되는 함수 ::############################################  
        def save_to_txt():
            if meaning == '' or spelling == '':
                #tkinter.messagebox.showwarning('알림','뜻과 스펠링을 전부 입력해주세요.') #얘는 메세지 박스 띄우는 애.
                save_word.config(text = '뜻과 스펠링을 전부 입력해주세요.')
            else:        
                f = open("C:/Users/HK/Capstone1/Words.txt", 'a')
                f.write(meaning + ' ' + spelling + '\n')
                f.close()
                save_word.config(text = '저장 완료')
                word_mean_enter.delete(0, tkinter.END)
                word_spelling_enter.delete(0, tkinter.END)
                check_mean.config(text = "입력한 뜻 : ")
                check_spelling.config(text = "입력한 스펠링 : ")
        
        ###########################################:: Button에 의해 실행되는 함수 ::############################################  
        

        
        #################################################:: Button 생성 ::#################################################
        #save_word = tkinter.Button(self, command = save_to_txt, text = '저장', anchor = 'center')
        #save_word.place(x=50, y=150, width=200)
        
        #tkinter.Button(self, command = save_to_txt, text = '저장', anchor = 'center')
        #tkinter.Button(self, command = save_to_txt, text = '저장', anchor = 'center').pack()
        save_word = tkinter.Button(self, command = save_to_txt, text = '저장', anchor = 'center')

        #tkinter.Button.pack(self, side="top")
        save_word.place(x=50, y=50)
        #################################################:: Button 생성 ::#################################################
        
        '''
        현재 문제점: 씨@발 위젯 배치가 안된다. 뒤에 .pack() 을 하는 이유도 모르겠다. 저거 하나 하고나면 가장 마지막에 설정한
        pack 설정에 따라 모든 버튼이 적용되어버린다.
        '''
        
        #tkinter.Frame.configure(self,bg='red')
        #tkinter.Label(self, text="Page two", font=('Helvetica', 18, "bold")).pack(side="top", fill="x", pady=5)
        
        back = tkinter.Button(self, text="←", command=lambda: master.switch_frame(StartPage))
        back.pack(side="top", anchor = 'w')        
       
        #################################################:: Entry 생성 ::#################################################
        word_mean_enter = tkinter.Entry(self)
        word_mean_enter.bind("<Return>", save_mean)
        word_mean_enter.place(x=100, y=45) 

        word_spelling_enter = tkinter.Entry(self)
        word_spelling_enter.bind("<Return>", save_spelling)
        word_spelling_enter.place(x=100, y=95) 
        #################################################:: Entry 생성 ::#################################################  
        
        #################################################:: Label 생성 ::#################################################
        enter_new_word = tkinter.Label(self, text = "단어의 뜻과 스펠링을 입력하세요.")
        enter_new_word.place(x=60, y=10)

        word_mean = tkinter.Label(self, text = "한글뜻 : ", anchor = 'e')
        word_mean.place(x=30, y=45, width=65)

        check_mean = tkinter.Label(self, text = "입력한 뜻 : ")
        check_mean.place(x=90, y=70)

        word_spelling = tkinter.Label(self, text = "스펠링 : ",  anchor = 'e')
        word_spelling.place(x=30, y=95, width=65)

        check_spelling = tkinter.Label(self, text = "입력한 스펠링 : ")
        check_spelling.place(x=90, y=120)
        #################################################:: Label 생성 ::#################################################
        
        

if __name__ == "__main__":
    print("main")
    app = SampleApp()
    app.mainloop()        
```
