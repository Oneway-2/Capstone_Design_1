```python
from tkinter import *
from tkinter.font import Font
import random
import time
```

```python
#global meaning, spelling

############################################:: word.txt 내용을 리스트로 담는 과정 ::###########################################  

word_txt, meaning_txt, spelling_txt = [], [] ,[]   # 미리 선언하지 않으면 out of range 나온다... 

def read_words():  # 리스트에 새롭게 입력된 단어 집어넣기
    f = open("C:/Users/HK/Capstone1/Words.txt", 'r')

    word_txt = f.readlines()
    print(len(word_txt))

    meaning_txt = [0 for i in range(len(word_txt))]
    spelling_txt = [0 for i in range(len(word_txt))]  # 이만큼 인덱스를 넓혀주지 않으면 out of range 나옴.

    for i in range(0, len(word_txt)):
        meaning_txt[i] = word_txt[i].split(' ')[0]   # meaning_txt 에 한글뜻이 전부 담긴다.
        spelling_txt[i] = word_txt[i].split(' ')[1]   #spelling_txt에 스펠링이 전부 담긴다.
    
    print('in read words function\n')
    print(meaning_txt)
    print(spelling_txt)
    f.close()
    return word_txt, meaning_txt, spelling_txt

word_txt, meaning_txt, spelling_txt = read_words()  # read_words 함수를 쓸 땐 꼭 이렇게 return 을 받자.

############################################:: word.txt 내용을 리스트로 담는 과정 ::###########################################  


############################################:: Entry에 의해 실행되는 함수 ::###########################################  

def save_mean(enter_value):   # 뜻을 입력받고 meaning 에 저장  # word_mean_enter Entry와 엮여있다.
    global meaning
    meaning = Entry.get(word_mean_enter)
    check_mean.config(text='입력한 뜻 : ' + meaning)
    print(meaning)
    save_word.config(text = '저장')
    
def save_spelling(enter_value):   # 스펠링을 입력받고 meaning 에 저장  # word_spelling_enter Entry와 엮여있다.
    global spelling
    spelling = Entry.get(word_spelling_enter)
    check_spelling.config(text='입력한 스펠링 : ' + spelling)
    print(spelling)
    save_word.config(text = '저장')

###########################################:: Entry에 의해 실행되는 함수 ::############################################  



###########################################:: Button에 의해 실행되는 함수 ::############################################  

def save_to_txt(word_txt, meaning_txt, spelling_txt):
    global meaning, spelling
    if meaning == '' or spelling == '':
        #tkinter.messagebox.showwarning('알림','뜻과 스펠링을 전부 입력해주세요.') #얘는 메세지 박스 띄우는 애.
        save_word.config(text = '뜻과 스펠링을 전부 입력해주세요.')
        print('in if function\n')
        print(meaning_txt)
        print(spelling_txt)
    elif meaning_txt.count(meaning) != 0 and spelling_txt.count(spelling+'\n') != 0:
        save_word.config(text = '이미 있는 단어&뜻 입니다.')
    else:        
        f = open("C:/Users/HK/Capstone1/Words.txt", 'a')
        f.write(meaning + ' ' + spelling + '\n')
        f.close()
        save_word.config(text = '저장 완료')
        word_mean_enter.delete(0, END)
        word_spelling_enter.delete(0, END)
        check_mean.config(text = "입력한 뜻 : ")
        check_spelling.config(text = "입력한 스펠링 : ")
        word_txt, meaning_txt, spelling_txt = read_words() # 세 리스트에 새롭게 추가된 단어까지 추가
        #### 함수 바깥에 있는 원래 리스트엔 갱신이 안된다.. 해결 필요!!!!!!!!!!!
        meaning, spelling = '', ''
        print('in save_to_txt function\n')
        print(meaning_txt)
        print(spelling_txt)
        
###########################################:: Button에 의해 실행되는 함수 ::############################################


def raise_frame(frame):
    frame.tkraise()
```

```python
root = Tk()
root.title('Helper')
root.geometry("300x200+1550+50")   # 창 크기 (너비x높이+x좌표+y좌표)
root.resizable(False, False)   # 창 크기 조절 여부

f1 = Frame(root, width = 300, height = 200)  # 암기모드
f2 = Frame(root, width = 300, height = 200)  # 시험모드
f3 = Frame(root, width = 300, height = 200)  # 입력모드
f4 = Frame(root, width = 300, height = 200)

for frame in (f1, f2, f3, f4):
    frame.grid(row=0, column=0, sticky='news')
    
#======================================================= f1 start =======================================================#  



#################################################:: Label 생성 ::#################################################

list_index = random.randint(0, len(word_txt) - 1)  # 0 부터 word_txt의 줄 수 만큼 랜덤한 수 저장

label_font=Font(family='Helvetica', size=20)

show_meaning = Label(f1, text = meaning_txt[list_index], anchor = 'center', width = 12, font = label_font)
show_meaning.place(x=50, y=30)

show_spelling = Label(f1, text = spelling_txt[list_index], width = 12, font = label_font)
show_spelling.place(x=50, y=100)

#################################################:: Label 생성 ::#################################################

#################################################:: Button 생성 ::#################################################

goto2 = Button(f1, text='시험 모드', command=lambda:raise_frame(f2))  # 2번으로 가는 버튼
goto2.place(x=1, y=174)
goto3 = Button(f1, text='입력 모드', command=lambda:raise_frame(f3))  # 3번으로 가는 버튼
goto3.place(x=237, y=174)

#################################################:: Button 생성 ::#################################################

 

#======================================================== f1 end ========================================================#  





#======================================================= f2 start =======================================================#  

Label(f2, text='FRAME 2').pack()
Button(f2, text='Go to frame 3', command=lambda:raise_frame(f1)).pack()

#======================================================== f2 end ========================================================#  





#======================================================= f3 start =======================================================#  

Button(f3, text='←', command=lambda:raise_frame(f1)).place(x=0, y=0)

#################################################:: Entry 생성 ::#################################################

word_mean_enter = Entry(f3)
word_mean_enter.bind("<Return>", save_mean)
word_mean_enter.place(x=100, y=45) 

word_spelling_enter = Entry(f3)
word_spelling_enter.bind("<Return>", save_spelling)
word_spelling_enter.place(x=100, y=95) 

#################################################:: Entry 생성 ::#################################################  

#################################################:: Label 생성 ::#################################################

enter_new_word = Label(f3, text = "단어의 뜻과 스펠링을 입력하세요.")
enter_new_word.place(x=60, y=10)

word_mean = Label(f3, text = "한글뜻 : ", anchor = 'e')
word_mean.place(x=30, y=45, width=65)

check_mean = Label(f3, text = "입력한 뜻 : ")
check_mean.place(x=90, y=70)

word_spelling = Label(f3, text = "스펠링 : ",  anchor = 'e')
word_spelling.place(x=30, y=95, width=65)

check_spelling = Label(f3, text = "입력한 스펠링 : ")
check_spelling.place(x=90, y=120)

#################################################:: Label 생성 ::#################################################

#################################################:: Button 생성 ::#################################################

save_word = Button(f3, command = lambda:save_to_txt(word_txt, meaning_txt, spelling_txt), text = '저장', anchor = 'center')
save_word.place(x=50, y=150, width=200)

#################################################:: Button 생성 ::#################################################

#======================================================== f3 end ========================================================#  





#======================================================= f4 start =======================================================#  

Label(f4, text='FRAME 4').pack()
Button(f4, text='Goto to frame 1', command=lambda:raise_frame(f1)).pack()

#======================================================== f4 end ========================================================#  



raise_frame(f1)  # 시작은 f1 부터
root.mainloop()
```
